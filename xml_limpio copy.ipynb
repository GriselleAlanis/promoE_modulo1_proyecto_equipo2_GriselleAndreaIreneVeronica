# Limpieza XML

## Imports

```python
import xml.etree.ElementTree as ET
```

## Lectura

```python
tree = ET.parse('data_xml.xml')
root = tree.getroot()

xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        if subchild.tag == "level_0":
            continue
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
xml_list
```

```python
def limpieza_xml(xml_list):
    
    copia_xml_list = xml_list

    # Llevar esto a una funcin
    for i in copia_xml_list:
        # i.pop('level_0') -> Lo hemos quitado de la lista xml_list. Ya no es necesario.
        if i.get('gender') == '0':
            i['gender'] = 'Man'

        elif i.get('gender') == '1':
            i['gender'] = 'Woman'

        elif i.get('gender') == '2':
            i['gender'] = 'Nonbinary'
        
        elif i.get('gender') == '3':
            i['gender'] = 'Prefer not to say'

        elif i.get('gender') == '4':
            i['gender'] = 'Prefer to self-describe'

        # Convertimos el value de 'index' a int
        i['index'] = int(i['index'])

    resultado = {}
    key = 1
    for i in copia_xml_list:
        tuple_values = tuple(i.values())
        resultado[key] = tuple_values
        key += 1        

    return resultado
```

```python
final_xml = limpieza_xml(xml_list)
final_xml
```
