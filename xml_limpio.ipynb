```python
import re
import os
import xml.etree.ElementTree as ET

tree = ET.parse('data_xml.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios
```

```python
xml_list
```

```python
def limpieza_xml(xml_list):
        
    for i in xml_list: 
        i.pop('level_0')
        if i.get('gender') == '0':
            i['gender'] = 'Man'

        elif i.get('gender') == '1':
            i['gender'] = 'Woman'

        elif i.get('gender') == '2':
            i['gender'] = 'Nonbinary'
        
        elif i.get('gender') == '3':
            i['gender'] = 'Prefer not to say'

        elif i.get('gender') == '4':
            i['gender'] = 'Prefer to self-describe'

    for i in xml_list:
        xml_automatizado = tuple(i.values())
        print(xml_automatizado)

    return xml_automatizado
```

```python
final_xml = limpieza_xml(xml_list)
final_xml
```
